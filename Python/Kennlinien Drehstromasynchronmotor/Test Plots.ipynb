```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
U20=230

kT=1/160

R2=0.2
X2streu=0.365
```

```python
def I2(s):
    return U20*s/np.sqrt(R2**2+(X2streu*s)**2)
```

```python
def cosphi2(s):
    return R2/np.sqrt((R2)**2+(s*X2streu)**2)
```

```python
def M(s):
    return kT*U20*I2(s)*cosphi2(s)
```

```python
x=np.arange(0,1,0.01)
y=M(x)
```

```python
fig = plt.figure()
ax=fig.subplots(1,1)

ax.plot(x,y)

ax.plot([0.03333], [98.78], marker='o', markersize=3, color="red") ## Nennmoment, Nennschlupf

ax.invert_xaxis()
ax.set_xlabel('Schlupf')
ax.set_ylabel('Drehmoment')
```

```python
R2Strich=0.0439
X2Strich=0.1523
```

```python
def MEdutecnica(s):
    return 2*R2Strich*(400/np.sqrt(3))**2/(2*np.pi*50*s*((0.2+R2Strich/s)**2)+(0.365+X2Strich)**2)
```
