```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
img = cv2.imread('Man.jpg',0)
```

```python
f = np.fft.fft2(img)
fshift =np.fft.fftshift(f)
magnitude_spectrum = 100*np.log(np.abs(fshift))
```

```python
fig, ax=plt.subplots(1,2,figsize=(20,10))
ax[0].imshow(img, cmap = 'gray')
ax[1].imshow(magnitude_spectrum,cmap="gray")
```

```python
rows, cols = img.shape
crow,ccol = rows//2 , cols//2
windowRange=30
fshift[crow-windowRange:crow+windowRange, ccol-windowRange:ccol+windowRange] *= 0
f_ishift = np.fft.ifftshift(fshift)

img_back = np.fft.ifft2(f_ishift)
img_back = np.abs(img_back)
```

```python
fig, ax=plt.subplots(2,2,figsize=(15,20))
ax[0,0].imshow(img, cmap = 'gray')
ax[0,1].imshow(magnitude_spectrum,cmap="gray")
ax[1,0].imshow(np.abs(fshift), cmap = 'gray')
ax[1,1].imshow(img_back)
```
